```python
import pandas as pd

# Importar o link do arquivo csv
uri ='https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/refs/heads/master/aula4.1/movies.csv'
# Adiciono a variavel filmes a função de read_csv,serve para ler arquivos csv
filmes = pd.read_csv(uri)
# Exibo os filmes porem apenas os 5 primeiros com a função head
filmes.head()
```

```python
# Uso a função columns para renoemar as colunas
filmes.columns = ['Filmes', 'Titulos', 'Generos']
filmes.head()
```
